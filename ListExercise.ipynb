## Alıştırmalar (Listeler)

1- Bir listedeki tüm öğelerin toplamını hesaplayan bir Python programı yazın.

```python
my_list = [1,2,3,4,5]
sum(my_list)
```

```python
total = 0
for i in my_list:
    total = total + i
    print(total)
```

2-Bir listedeki en büyük ve en küçük sayıyı bulan bir Python programı yazın.

```python
minimum = my_list[0]
maximum = my_list[0]
for number in my_list:
    if number > maximum:
        maximum = number
        
    if number < minimum:
        minimum = number

print("maksimum is: " ,maximum)        
print("minimum is: " ,minimum)
```

```python
min(my_list)
```

```python
max(my_list)
```
