1- Üç sayının en büyüğünü bulan bir Python fonksiyonu yazın.

İpucu: Önce iki sayının en büyüğünü bulan bir fonksiyon yazın.
Daha sonra bu fonksiyonu kullanarak üç sayının en büyüğünü bulun.

```python
def maximum(a,b,c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c
```

```python
print(maximum(3,5,2))
```

2- Bir listedeki tüm sayıların çarpımını hesaplayan bir Python fonksiyonu yazın.

İpucu: Bir for döngüsü kullanın.

```python
def multiplication(my_list):
    multi = 1

    for i in my_list:
        multi = multi * i
    return multi
```

```python
sample_list = [8, 2, 3, -1, 7]
multiplication(sample_list)
```

3- Bir liste alan ve bu listenin benzersiz (unique) elemanlarından oluşan yeni bir liste döndüren Python fonksiyonu yazın.

İpucu 1: for döngüsü ve if ifadesini birlikte kullanın.
İpucu 2: count() metodu, bir listede belirli bir elemanın kaç kez geçtiğini sayar.

```python
def unique_element(my_list):
    output = set(my_list) #Küme elemanları kategorik olarak unique (benzersiz) olmak zorunda, o yüzden liste yerine küme tercih edildi.
    return output
```

```python
example_list = [1,1,3,3,3,3,4,4,4,5,7,7,7]
unique_element(example_list) #Beklenen çıktı küme (süslü parantez) şeklinde olacak, bunu tekrar listeye çevirmeliyiz.
```

```python
list(unique_element(example_list)) # kümenin listeye çevrilmiş hali.
```

```python
#2.Çözüm
def unique(example_list):
    copy_list = example_list.copy()
    output = []
    for i in example_list:
        if copy_list.count(i) > 1:
            copy_list.remove(i)
        else:
            output.append(i)
    return output
```

```python
unique(example_list)
```

4- Bir sayı parametresi alan ve sayının asal olup olmadığını kontrol eden bir Python fonksiyonu yazın.

İpucu:
Asal sayı, 1’den büyük ve kendisi ile 1 dışında pozitif böleni olmayan doğal sayıdır.

```python
def asal_mi(x):
    asal_mi = True

    if type(x) == float or x <= 1:   # x float ise veya 1'den küçük/eşitse
        asal_mi = "Tanımsız"
    else:
        for i in range(2, x):        # 2'den x-1'e kadar dene
            if x % i == 0:           # tam bölünüyorsa
                asal_mi = False
                break

    return asal_mi
```

```python
asal_mi(3)
```
