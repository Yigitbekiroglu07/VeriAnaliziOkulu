1- Bir sayı dizisindeki çift ve tek sayıların sayısını hesaplayan bir Python programı yazın.

```python
sample_list = [2, 3, 4, 5, 8, 2, 9]
single_number = []
even_number = []
for x in sample_list:
    if x % 2 == 0:
        even_number.append(x)
    else:
        single_number.append(x)

print(len(even_number))
print(len(single_number))
```

2- 0 ile 50 arasındaki Fibonacci dizisini bulan bir Python programı yazın.

İpucu: Fibonacci Dizisi şu şekildedir:
0, 1, 1, 2, 3, 5, 8, 13, 21, ...
Her bir sonraki sayı, kendisinden önceki iki sayının toplamıdır.

```python
fib = [0, 1]
for f in range(0,50,1):
    if f == fib[-1] + fib[-2]:
        fib.append(f)
    else:
        continue
print(fib)
```

3.c. 10'dan küçük tamsayılardan oluşan bir liste oluşturun ve adını a koyun.

```python
x = [1,3,15,6]
total = 0
for i in x:
    total = total + i
print(total)
```

```python
x = [1,3,15,6]
minimum = x[0] # x dizisindeki herhangi bir değeri versek olur. Boş önerme - null hipotez
maximum = x[0] # x dizisindeki herhangi bir değeri versek olur. Boş önerme - null hipotez
for i in x:
    if i < minimum:
        minimum = i
    elif i > maximum:
        maximum = i
print(minimum)
print(maximum)
```
