1- Movie sözlüğündeki "name" anahtarının değerini get metodu ile yazdırın.

```python
Movie = {
  "name": "Twin Peaks",
  "director": "David Lynch",
  "year": 1990
}
```

```python
Movie.get("name")
```

2- Year anahtarını 1990'dan 1991'e değiştirin.

```python
Movie["year"] = 1991
```

```python
Movie
```

```python
Movie.update({"genre": "mystery/drama/horror"})
```
