```python
a = np.random.random(1000)
idx = np.array([])

for i in range(len(a)):
    ele = a[i]
    if ele > 0.5:
        idx = np.append(idx,i)

print(len(idx))
```

Repeat fonksiyonu nasıl çalışır?

```python
print(np.repeat(3, 4))              # 3'ü 4 kez tekrarlar

x = np.array([[1,2],[3,4]])         # 2x2 matris
print(x)                            # matrisi yazdırır

print(np.repeat(x, 2))              # tüm elemanları (düzleştirip) 2 kez tekrarlar
print(np.repeat(x, 3, axis=0))      # satırları 3 kez tekrarlar
print(np.repeat(x, 3, axis=1))      # sütunları 3 kez tekrarlar
print(np.repeat(x, [1, 2], axis=0)) # 1. satır 1 kez, 2. satır 2 kez tekrar
```

12- Belirli bir dizi (dizi) ve istediğiniz bir adım (veya pencere boyutu) için bir time_resolved fonksiyonu yazın.

Fonksiyon şu üç parametreyi alacak:
time_resolved(dizi, pencere_genişliği, kayma)

Örnek olarak,
dizi = np.array([1, 2, 6, 4, 5, 4, 2, 7, 8, 9]) için
time_resolved(dizi, 3, 4) çağrıldığında, fonksiyonun çıktısı [4, 5, 4] olmalıdır.

```python
def time_resolved(dizi,pencere_genişliği,kayma):
    alt_dizi=np.array([])
    for i in range(kayma-1,kayma+pencere_genişliği-1):
        alt_dizi=np.append(alt_dizi,dizi[i])
    return alt_dizi
```

```python
dizi = np.array([1, 2, 6, 4, 5, 4, 2, 7, 8, 9])
time_resolved(dizi, 3, 4)
```
